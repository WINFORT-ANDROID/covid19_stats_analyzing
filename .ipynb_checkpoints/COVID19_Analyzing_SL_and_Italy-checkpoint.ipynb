```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
confirmed_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
```

```python
italy_confirmed_df = confirmed_df.loc[confirmed_df['Country/Region'] == 'Italy']
sl_confirmed_df = confirmed_df.loc[confirmed_df['Country/Region'] == 'Sri Lanka']
```

```python
x_axis = confirmed_df.columns[-7:]
x_axis
```

```python
italy_data = []
for index, row in italy_confirmed_df.iterrows():
    italy_data = row[-7: ].values
italy_data
```

```python
sl_data = []
for index, row in sl_confirmed_df.iterrows():
    sl_data = row[-7: ].values
sl_data
```

```python
x_indexes = np.arange(len(x_axis))
width = 0.25
plt.figure(figsize=(10, 5), dpi=100)

plt.plot(x_axis, italy_data, label='Italy', color='red', marker='.', linewidth=2, markersize=10, markeredgecolor='black')
plt.plot(x_axis, sl_data, label='SL', color='blue', linestyle='--', linewidth=2, marker='.', markersize=10, markeredgecolor='black')

plt.title('COVID-19')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')

plt.legend()
plt.style.use('fivethirtyeight')
# plt.grid()
plt.tight_layout()

# plt.savefig('confirmed.png', dpi=200)

plt.show()
```
