```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_confirmed = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
df_death = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
df_recovered = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
```

```python
df_confirmed_china = df_confirmed.loc[df_confirmed['Country/Region'] == 'China']
df_recovered_china = df_recovered.loc[df_recovered['Country/Region'] == 'China']
df_death_china = df_death.loc[df_death['Country/Region'] == 'China']
```

```python
confirmed_data_china = df_confirmed_china.iloc[:, 4:].sum(axis=0, skipna=True).values
recovered_data_china = df_recovered_china.iloc[:, 4:].sum(axis=0, skipna=True).values
death_data_china = df_death_china.iloc[:, 4:].sum(axis=0, skipna=True).values
active_data_china = confirmed_data_china - (recovered_data_china + death_data_china)
```

```python
plt.figure(figsize=(7, 4), dpi=100)

plt.plot(df_confirmed_china.columns[4: ], confirmed_data_china, label='Confirmed Cases', color='blue', linewidth=2, marker='.', markersize=10)
plt.plot(df_recovered_china.columns[4: ], recovered_data_china, label='Recovered Cases', color='#049372', linewidth=2, marker='.', markersize=10)
plt.plot(df_death_china.columns[4: ], death_data_china, label='Deaths', color='red', linewidth=2, marker='.', markersize=10)
plt.plot(df_death_china.columns[4: ], active_data_china, label='Active Cases', color='green', linewidth=2, marker='.', markersize=10)

plt.title('COVID19 Outbreak in China')
plt.xlabel('date')
plt.ylabel('cases reported')
plt.xticks(df_confirmed_china.columns[8::8])

plt.tight_layout()
plt.grid()
plt.legend()

plt.show()
```
