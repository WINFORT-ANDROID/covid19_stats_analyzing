```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_confirmed_USA = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv', index_col='UID')
pd.set_option('display.max_columns', 83)
pd.set_option('display.max_rows', 3253)

# df_confirmed_USA
```

```python
df_death_USA = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_US.csv', index_col='UID')
# df_death_USA
```

```python
df_recovered_USA = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
```

```python
plt.figure(figsize=(7, 4), dpi=100)

plt.bar(df_confirmed_USA.columns[10: ], df_confirmed_USA.iloc[:, 10:].sum(axis=0, skipna=True).values, label='Confirmed Cases', color='blue')
plt.bar(df_recovered_USA.columns[4: ], df_recovered_USA.loc[df_recovered_USA['Country/Region'] == 'US'].iloc[0, 4:].values, label='Recovered Cases', color='green')
plt.bar(df_death_USA.columns[11: ], df_death_USA.iloc[:, 11: ].sum(axis=0, skipna=True).values, label='Deaths', color='red')


# plt.plot(df_confirmed_USA.columns[10: ], df_confirmed_USA.iloc[:, 10:].sum(axis=0, skipna=True).values, label='Confirmed Cases', linewidth=2, marker='.', color='blue', markeredgecolor='blue', markersize=10)
# plt.plot(df_recovered_USA.columns[4: ], df_recovered_USA.loc[df_recovered_USA['Country/Region'] == 'US'].iloc[0, 4:].values, label='Recovered Cases', linewidth=2, marker='.', color='green', markeredgecolor='green', markersize=10)
# plt.plot(df_death_USA.columns[11: ], df_death_USA.iloc[:, 11: ].sum(axis=0, skipna=True).values, label='Death Cases', linewidth=2, marker='.', color='red', markeredgecolor='red', markersize=10)


plt.title('COVID-19 Outbreak in China')
plt.xticks(df_confirmed_USA.columns[10::10])

plt.grid()
plt.legend()
plt.tight_layout()

plt.show()
```
