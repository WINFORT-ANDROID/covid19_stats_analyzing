```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_recovered = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
pd.set_option('display.max_columns', 70)
```

```python
df_china = df_recovered.loc[df_recovered['Country/Region'] == 'China']
recovered_total_china = df_china.sum(axis=0, skipna=True)[4: ].values
# recovered_total_china
```

```python
df_us = df_recovered.loc[df_recovered['Country/Region'] == 'US']
recovered_total_us = df_us.sum(axis=0, skipna=True)[4: ].values
# recovered_total_us
```

```python
df_italy = df_recovered.loc[df_recovered['Country/Region'] == 'Italy']
recovered_total_italy = df_italy.sum(axis=0, skipna=True)[4: ].values
# recovered_total_italy
```

```python
df_spain = df_recovered.loc[df_recovered['Country/Region'] == 'Spain']
recovered_total_spain = df_spain.sum(axis=0, skipna=True)[4: ].values
# recovered_total_spain
```

```python
df_iran = df_recovered.loc[df_recovered['Country/Region'] == 'Iran']
recovered_total_iran = df_iran.sum(axis=0, skipna=True)[4: ].values
# recovered_total_iran
```

```python
df_uk = df_recovered.loc[df_recovered['Country/Region'] == 'United Kingdom']
recovered_total_uk = df_uk.sum(axis=0, skipna=True)[4: ].values
# recovered_total_uk
```

```python
df_france = df_recovered.loc[df_recovered['Country/Region'] == 'France']
recovered_total_france = df_france.sum(axis=0, skipna=True)[4: ].values
# recovered_total_france
```

```python
recovered_total_date = df_france.columns[4:]
# recovered_total_date
```

```python
plt.figure(figsize=(8, 5), dpi=100)
plt.plot(recovered_total_date, recovered_total_china, label='China', linewidth=2, marker='.', markersize=8, color='#C93756', markeredgecolor='#C93756')
plt.plot(recovered_total_date, recovered_total_italy, label='Italy', linewidth=2, marker='.', markersize=8, color='#8E44AD', markeredgecolor='#8E44AD')
plt.plot(recovered_total_date, recovered_total_spain, label='Spain', linewidth=2, marker='.', markersize=8, markeredgecolor='#FFA631', color='#FFA631')
# plt.plot(recovered_total_date, recovered_total_uk, label='UK', linewidth=2, marker='.', markersize=8, markeredgecolor='m', color='m')
plt.plot(recovered_total_date, recovered_total_iran, label='Iran', linewidth=2, marker='.', markersize=8, markeredgecolor='#049372', color='#049372')
plt.plot(recovered_total_date, recovered_total_us, label='US', linewidth=2, marker='.', markersize=8, markeredgecolor='#1F4788', color='#1F4788')
# plt.plot(recovered_total_date, recovered_total_france, label='France', linewidth=2, marker='.', markersize=8, markeredgecolor='gray', color='gray')

plt.title('COVID-19 Recovered Cases', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 15})
plt.xlabel('date', fontdict={'fontname': 'Comic Sans MS'})
plt.ylabel('cases re', fontdict={'fontname': 'Comic Sans MS'})
plt.xticks(recovered_total_date[4::7])

# plt.tight_layout()
plt.grid()
plt.legend()

plt.show()
```
