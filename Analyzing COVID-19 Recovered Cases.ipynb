```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_recovered = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
pd.set_option('display.max_columns', 70)
```

```python
df_china = df_recovered.loc[df_recovered['Country/Region'] == 'China']
recovered_total_china = df_china.sum(axis=0, skipna=True)[4: ].values[::10]
# recovered_total_china
```

```python
df_us = df_recovered.loc[df_recovered['Country/Region'] == 'US']
recovered_total_us = df_us.sum(axis=0, skipna=True)[4: ].values[::10]
# recovered_total_us
```

```python
df_italy = df_recovered.loc[df_recovered['Country/Region'] == 'Italy']
recovered_total_italy = df_italy.sum(axis=0, skipna=True)[4: ].values[::10]
# recovered_total_italy
```

```python
df_spain = df_recovered.loc[df_recovered['Country/Region'] == 'Spain']
recovered_total_spain = df_spain.sum(axis=0, skipna=True)[4: ].values[::10]
# recovered_total_spain
```

```python
df_iran = df_recovered.loc[df_recovered['Country/Region'] == 'Iran']
recovered_total_iran = df_iran.sum(axis=0, skipna=True)[4: ].values[::10]
# recovered_total_iran
```

```python
df_uk = df_recovered.loc[df_recovered['Country/Region'] == 'United Kingdom']
recovered_total_uk = df_uk.sum(axis=0, skipna=True)[4: ].values[::10]
# recovered_total_uk
```

```python
df_france = df_recovered.loc[df_recovered['Country/Region'] == 'France']
recovered_total_france = df_france.sum(axis=0, skipna=True)[4: ].values[::10]
# recovered_total_france
```

```python
recovered_total_date = df_france.columns[4::10]
# recovered_total_date
```

```python
plt.figure(figsize=(8, 4), dpi=100)
plt.plot(recovered_total_date, recovered_total_china, label='China', linewidth=2, marker='.', markersize=8, markeredgecolor='red')
plt.plot(recovered_total_date, recovered_total_italy, label='Italy', linestyle=':', linewidth=2, marker='.', markersize=8, markeredgecolor='purple')
plt.plot(recovered_total_date, recovered_total_spain, label='Spain', linestyle='-.', linewidth=2, marker='.', markersize=8, markeredgecolor='green', color='m')
plt.plot(recovered_total_date, recovered_total_uk, label='UK', linestyle='-.', linewidth=2, marker='.', markersize=8, markeredgecolor='yellow', color='black')
plt.plot(recovered_total_date, recovered_total_iran, label='Iran', linestyle=':', linewidth=2, marker='.', markersize=8, markeredgecolor='black', color='green')
plt.plot(recovered_total_date, recovered_total_us, label='US', linewidth=2, marker='.', markersize=8, markeredgecolor='blue', color='cyan')

plt.title('COVID-19 Recovered Cases', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('date', fontdict={'fontname': 'Comic Sans MS'})
plt.ylabel('cases by last 10 days', fontdict={'fontname': 'Comic Sans MS'})

# plt.tight_layout()
plt.grid()
plt.legend()

plt.show()
```
