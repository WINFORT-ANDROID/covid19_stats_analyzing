```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_confirmed = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
# df_confirmed = pd.read_csv('./data/covid19_confirmed.csv')
pd.set_option('display.max_columns', 70)
```

```python
df_china = df_confirmed.loc[df_confirmed['Country/Region'] == 'China']
confirmed_total_china = df_china.sum(axis=0, skipna=True)[4: ].values[::10]
# confirmed_total_china
```

```python
df_us = df_confirmed.loc[df_confirmed['Country/Region'] == 'US']
confirmed_total_us = df_us.sum(axis=0, skipna=True)[4: ].values[::10]
# confirmed_total_us
```

```python
df_italy = df_confirmed.loc[df_confirmed['Country/Region'] == 'Italy']
confirmed_total_italy = df_italy.sum(axis=0, skipna=True)[4: ].values[::10]
# confirmed_total_italy
```

```python
df_spain = df_confirmed.loc[df_confirmed['Country/Region'] == 'Spain']
confirmed_total_spain = df_spain.sum(axis=0, skipna=True)[4: ].values[::10]
# confirmed_total_spain
```

```python
df_iran = df_confirmed.loc[df_confirmed['Country/Region'] == 'Iran']
confirmed_total_iran = df_iran.sum(axis=0, skipna=True)[4: ].values[::10]
# confirmed_total_iran
```

```python
df_uk = df_confirmed.loc[df_confirmed['Country/Region'] == 'United Kingdom']
confirmed_total_uk = df_uk.sum(axis=0, skipna=True)[4: ].values[::10]
# confirmed_total_uk
```

```python
df_france = df_confirmed.loc[df_confirmed['Country/Region'] == 'France']
confirmed_total_france = df_france.sum(axis=0, skipna=True)[4: ].values[::10]
confirmed_total_france
```

```python
confirmed_total_date = df_uk.columns[4::10]
confirmed_total_date
```

```python
plt.figure(figsize=(8, 4), dpi=100)
plt.plot(confirmed_total_date, confirmed_total_china, label='China', linewidth=2, marker='.', markersize=8, markeredgecolor='red')
plt.plot(confirmed_total_date, confirmed_total_italy, label='Italy', linestyle=':', linewidth=2, marker='.', markersize=8, markeredgecolor='purple')
plt.plot(confirmed_total_date, confirmed_total_spain, label='Spain', linestyle='-.', linewidth=2, marker='.', markersize=8, markeredgecolor='green', color='m')
plt.plot(confirmed_total_date, confirmed_total_uk, label='UK', linestyle='-.', linewidth=2, marker='.', markersize=8, markeredgecolor='yellow', color='black')
plt.plot(confirmed_total_date, confirmed_total_iran, label='Iran', linestyle=':', linewidth=2, marker='.', markersize=8, markeredgecolor='black', color='green')
plt.plot(confirmed_total_date, confirmed_total_us, label='US', linewidth=2, marker='.', markersize=8, markeredgecolor='blue', color='cyan')
plt.plot(confirmed_total_date, confirmed_total_france, label='France', linewidth=2, marker='.', markersize=8, markeredgecolor='red', color='grey')


plt.title('COVID-19 Confirmed Cases', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('date', fontdict={'fontname': 'Comic Sans MS'})
plt.ylabel('cases by last 10 days', fontdict={'fontname': 'Comic Sans MS'})

plt.tight_layout()
plt.grid()
plt.legend()

plt.show()
```
